```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
veri = pd.read_csv("./data/olimpiyatlar_temizlenmis.csv")
veri.head(5)
```

```python
# Saçılım Grafiği, iki farklı değişkenin arasındaki ilişkiyi belirlemek için kullanılır.
def sacilimGrafikCiz():
  sns.scatterplot(x="boy", y="kilo", data=veri)
  plt.title("Boy ve Ağırlık İlişkisi - Beyaz Izgara Tema")
  plt.show()

sacilimGrafikCiz()
```

```python
sns.set_style("dark")
sacilimGrafikCiz()
```

```python
# Saçılım Grafiği - Kategorik
sns.scatterplot(x="boy", y="kilo", hue="madalya", data=veri)
plt.title("Madalyaya Göre Boy ve Ağırlık Dağılımı")
plt.show()
```

```python
sns.set_style("whitegrid")

sns.scatterplot(x="boy", y="kilo", hue="cinsiyet", style="madalya", data=veri)
plt.xlabel('Sporcuların Boyu')
plt.ylabel('Sporcuların Ağırlığı')
plt.title('Madalyalı Sporcuların Ağırlık-Boy Değişimleri')
plt.show()
```

```python
sns.set_style("darkgrid")

sns.scatterplot(x="boy", y="kilo", hue="cinsiyet", style='madalya', size='yas', sizes=(15,200), data=veri)
plt.ylabel("Sporcuların Boyu")
plt.xlabel("Sporcuların Kilosu")
plt.title("Madalyualı Sporcuların Ağırlık-Boy Değişimleri")
plt.show()


```

```python
# Saçılım Grafiği - Doğrusal Regresyon 

sns.regplot(x="boy", y="kilo", data=veri, marker="+", scatter_kws={'alpha': 0.2})
plt.title("Boy ve Ağırlık Dağılımı")
plt.show()
```

```python
# Renk Paletleri

sns.scatterplot(x="boy", y="kilo", hue="madalya", data=veri, palette="Set1")
plt.title("Madalyaya Göre Boy ve Ağırlık Dağılımı")

plt.show()
```

```python
sns.scatterplot(x="boy", y="kilo", hue="madalya", data=veri)
plt.title("Madalyaya Göre Boy ve Ağırlık Dağılımı")

plt.show()
```

```python
# Histogram,

sns.displot(veri, x='kilo')
plt.ylabel("Frekans")
plt.title("Ağırlık Histogramı")
plt.show()
```

```python
sns.displot(veri, x="kilo", hue="cinsiyet")
plt.ylabel("Frekans")
plt.title("cinsiyete göre ağırlık histogramı")
plt.show()
```

```python
sns.displot(veri, x="kilo", col="cinsiyet", multiple="dodge")
plt.show()
```

```python
# İki Boyutlu Histogram

sns.displot(veri, x="kilo", y="boy", kind="kde")
plt.xlabel("Ağırlık")
plt.ylabel("Boy")
plt.title("Ağırlık Boy Histogramı")
plt.show()
```

```python
sns.displot(veri, x="kilo", y="boy", hue="cinsiyet", kind="kde")
plt.xlabel("Ağırlık")
plt.ylabel("Boy")
plt.title("Cinsiyete göre Ağırlık Boy Histogramı")
plt.show()
```

```python
 sns.barplot(x="madalya", y="boy", data=veri, hue="cinsiyet")
 plt.title("madalyalara göre boy grafikleri")
 plt.show()
```

```python
sns.catplot(x="madalya", y="yas", hue="cinsiyet", col="sezon", data=veri, kind="bar", height=3, aspect=1)
plt.show()
```

```python
# KUTU GRAFİĞİ
sns.boxplot(x="sezon", y="boy", data=veri)
plt.show()
```

```python
sns.boxplot(x="sezon", y="boy", hue="cinsiyet", data=veri, palette="rocket")
plt.show()
```

```python
veri_gecici = veri.loc[:,["yas", "kilo", "boy"]]
sns.boxplot(data=veri_gecici, orient="h", palette="Set2")
plt.show()
```

```python
sns.catplot(x="sezon", y="boy", hue="cinsiyet", col="madalya",data=veri, kind="box", height=6 , aspect=1)
plt.show()
```

```python
sns.heatmap(veri_gecici.corr(), annot=True, linewidths=0.5, fmt='.1f')
plt.show()
```

```python
# KEMAN GRAFİĞİ
sns.violinplot(x="sezon", y="boy", hue="cinsiyet", data=veri, split=True)
plt.show()
```

```python
sns.catplot(x="sezon", y="boy", hue="cinsiyet", col="madalya",
             data=veri, kind="violin", split=True, height=4, aspect=0.7)
```

```python
g = sns.JointGrid(data=veri, x="kilo", y="boy")
g.plot_joint(sns.histplot)
g.plot_marginals(sns.boxplot)
plt.show()
```

```python
# Çift / Eş Grafiği

sns.pairplot(veri)
plt.show()
```

```python
g  = sns.PairGrid(veri)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot,fill=True)
g.map_diag(sns.crayons, kde=True)
```
