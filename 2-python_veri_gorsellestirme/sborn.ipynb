```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
veri = pd.read_csv("./data/olimpiyatlar_temizlenmis.csv")
veri.head(5)
```

```python
# Salm Grafii, iki farkl deikenin arasndaki ilikiyi belirlemek iin kullanlr.
def sacilimGrafikCiz():
  sns.scatterplot(x="boy", y="kilo", data=veri)
  plt.title("Boy ve Arlk likisi - Beyaz Izgara Tema")
  plt.show()

sacilimGrafikCiz()
```

```python
sns.set_style("dark")
sacilimGrafikCiz()
```

```python
# Salm Grafii - Kategorik
sns.scatterplot(x="boy", y="kilo", hue="madalya", data=veri)
plt.title("Madalyaya Gre Boy ve Arlk Dalm")
plt.show()
```

```python
sns.set_style("whitegrid")

sns.scatterplot(x="boy", y="kilo", hue="cinsiyet", style="madalya", data=veri)
plt.xlabel('Sporcularn Boyu')
plt.ylabel('Sporcularn Arl')
plt.title('Madalyal Sporcularn Arlk-Boy Deiimleri')
plt.show()
```

```python
sns.set_style("darkgrid")

sns.scatterplot(x="boy", y="kilo", hue="cinsiyet", style='madalya', size='yas', sizes=(15,200), data=veri)
plt.ylabel("Sporcularn Boyu")
plt.xlabel("Sporcularn Kilosu")
plt.title("Madalyual Sporcularn Arlk-Boy Deiimleri")
plt.show()


```

```python
# Salm Grafii - Dorusal Regresyon 

sns.regplot(x="boy", y="kilo", data=veri, marker="+", scatter_kws={'alpha': 0.2})
plt.title("Boy ve Arlk Dalm")
plt.show()
```

```python
# Renk Paletleri

sns.scatterplot(x="boy", y="kilo", hue="madalya", data=veri, palette="Set1")
plt.title("Madalyaya Gre Boy ve Arlk Dalm")

plt.show()
```

```python
sns.scatterplot(x="boy", y="kilo", hue="madalya", data=veri)
plt.title("Madalyaya Gre Boy ve Arlk Dalm")

plt.show()
```

```python
# Histogram,

sns.displot(veri, x='kilo')
plt.ylabel("Frekans")
plt.title("Arlk Histogram")
plt.show()
```

```python
sns.displot(veri, x="kilo", hue="cinsiyet")
plt.ylabel("Frekans")
plt.title("cinsiyete gre arlk histogram")
plt.show()
```

```python
sns.displot(veri, x="kilo", col="cinsiyet", multiple="dodge")
plt.show()
```

```python
# ki Boyutlu Histogram

sns.displot(veri, x="kilo", y="boy", kind="kde")
plt.xlabel("Arlk")
plt.ylabel("Boy")
plt.title("Arlk Boy Histogram")
plt.show()
```

```python
sns.displot(veri, x="kilo", y="boy", hue="cinsiyet", kind="kde")
plt.xlabel("Arlk")
plt.ylabel("Boy")
plt.title("Cinsiyete gre Arlk Boy Histogram")
plt.show()
```

```python
 sns.barplot(x="madalya", y="boy", data=veri, hue="cinsiyet")
 plt.title("madalyalara gre boy grafikleri")
 plt.show()
```

```python
sns.catplot(x="madalya", y="yas", hue="cinsiyet", col="sezon", data=veri, kind="bar", height=3, aspect=1)
plt.show()
```

```python
# KUTU GRAF
sns.boxplot(x="sezon", y="boy", data=veri)
plt.show()
```

```python
sns.boxplot(x="sezon", y="boy", hue="cinsiyet", data=veri, palette="rocket")
plt.show()
```

```python
veri_gecici = veri.loc[:,["yas", "kilo", "boy"]]
sns.boxplot(data=veri_gecici, orient="h", palette="Set2")
plt.show()
```

```python
sns.catplot(x="sezon", y="boy", hue="cinsiyet", col="madalya",data=veri, kind="box", height=6 , aspect=1)
plt.show()
```

```python
sns.heatmap(veri_gecici.corr(), annot=True, linewidths=0.5, fmt='.1f')
plt.show()
```

```python
# KEMAN GRAF
sns.violinplot(x="sezon", y="boy", hue="cinsiyet", data=veri, split=True)
plt.show()
```

```python
sns.catplot(x="sezon", y="boy", hue="cinsiyet", col="madalya",
             data=veri, kind="violin", split=True, height=4, aspect=0.7)
```

```python
g = sns.JointGrid(data=veri, x="kilo", y="boy")
g.plot_joint(sns.histplot)
g.plot_marginals(sns.boxplot)
plt.show()
```

```python
# ift / E Grafii

sns.pairplot(veri)
plt.show()
```

```python
g  = sns.PairGrid(veri)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot,fill=True)
g.map_diag(sns.crayons, kde=True)
```
